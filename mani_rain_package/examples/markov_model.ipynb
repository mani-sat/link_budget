```python
import mani_rain
import numpy as np
import matplotlib.pyplot as plt
```

Access model directly


```python
rain = []
for i in range(10000):
    rain.append(mani_rain.rain.aau_model.draw_rain())
plt.plot(rain)
plt.xlabel("# Sample")
plt.ylabel("Rain mmhr")
```

```python
rain_model = mani_rain.markov_rain(mani_rain.new_norcia,
                                   mani_rain.rain.nn_model)
attenuation = []
elevation = 50
for i in range(10000):
    attenuation.append(rain_model.attenuation_saunders(elevation))

plt.plot(attenuation)
plt.xlabel("# Sample")
plt.ylabel("Rain attenuation [dB]")
```

Add a link

```python
link = mani_rain.link_budget_markov(mani_rain.new_norcia, 
                                    rain_model,
                                    100e6)

link.snr_at_t(384400e3, 10)
snr = []
for _ in range(10000):
    snr.append(link.snr_at_t(384400e3, 45))
plt.plot(snr)
```
